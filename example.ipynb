```python
from TSMetrics import * 
```

```python
import numpy as np
M = 100
x = np.linspace(-np.pi,np.pi,M)
y_true = 3.4*np.sin(x)
N = 10**4
y_pred = np.array([np.random.choice(np.linspace(2.4,3.8,M))*np.sin(x) for _ in range(N)])
ts_metrics = TSMetrics(y_true,y_pred)
```

```python
data = ts_metrics.compute_standard_metrics_summary()
```
